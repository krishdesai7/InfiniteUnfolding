```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy.stats import wasserstein_distance

from unfold import *
from scripting import inf_unfold
```

```python
data_size = 10**4
sim_size = 10**5
mu_true, mu_gen = 0, -1
scale = 20

rng = np.random.default_rng()
```

```python
truth = rng.normal(mu_true, 1, (data_size, 1))
data = truth * scale
gen = rng.normal(mu_gen, 1, (sim_size, 1))
sim = gen * scale
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].hist(truth, alpha = 0.5, label = 'Truth', density = True)
ax[0].hist(gen, alpha = 0.5, label = 'Gen.', density = True)
ax[0].legend()
ax[0].set_xlim(-40, 40)

ax[1].hist(data, alpha = 0.5, label = 'Data', density = True)
ax[1].hist(sim, alpha = 0.5, label = 'Sim.', density = True)
ax[1].legend()

plt.show()
```

```python
x_det = np.concatenate((data, sim))
x_part = np.concatenate((truth, gen))
y = np.concatenate((np.ones(data_size), np.zeros(sim_size)))

x_det_train, x_det_test, x_part_train, x_part_test, y_train, y_test= train_test_split(x_det, x_part, y)
```

```python
iterations = 5
dims = 1
```

```python
weights, models, history, model = multifold(dims, iterations,
                                     x_part_train[y_train == 0], x_det_train[y_train==0], x_det_train[y_train==1], 
                                     verbose=0
                                    )
```

```python
weights_omnifold = model.predict(x_part_test)
```

```python
weights_inf_unfold = inf_unfold.main(x_det_train, x_part_train, y_train, x_part_test, y_test)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(5, 7), gridspec_kw={'height_ratios': [4, 2]})

bins = np.linspace(-3, 3, 30)
bin_centers = 0.5 * (bins[:-1] + bins[1:])
t, _, _ = ax[0].hist(x_part_test[y_test == 1], alpha = 0.5, density = True, label = 'Truth', bins = bins)
g, _, _, = ax[0].hist(x_part_test[y_test == 0], alpha = 0.5, density = True, label = 'Generation', bins = bins)
of, _, _ = ax[0].hist(x_part_test[y_test == 0], weights = weights_omnifold[y_test == 0], alpha = 0.5, density = True, color='black', histtype="step",
           label="OmniFold",lw=2, bins = bins
          )
inf_unf, _, _ = ax[0].hist(x_part_test[y_test == 0], weights = weights_inf_unfold, alpha = 0.5, density = True, color='red', histtype="step",
           label="InfUnfold",lw=4, ls = "--", bins = bins
          )
ax[0].legend()

ax[1].plot(bin_centers, g/t, color = 'orange', marker = 'o', ls = '--')
ax[1].plot(bin_centers, of/t, color = 'black', marker = '^', ls = '--')
ax[1].plot(bin_centers, inf_unf/t, color = 'red', marker = 'v', ls = '--')
ax[1].set_xlim(-2.5, 3)
ax[1].set_ylim(0, 2.5)

plt.show()
```

```python
print(f"Distance between Truth and")
print(f"\tGen = {wasserstein_distance(x_part_test[y_test == 1].squeeze(), x_part_test[y_test == 0].squeeze()):.4f}")
print(f"\tOmnifold = {wasserstein_distance(x_part_test[y_test == 1].squeeze(), x_part_test[y_test == 0].squeeze(), 
                                           v_weights = weights_omnifold[y_test == 0]):.4f}")
print(f"\tInf Unfold = {wasserstein_distance(x_part_test[y_test == 1].squeeze(), x_part_test[y_test == 0].squeeze(), 
                                           v_weights = weights_inf_unfold):.4f}")
```
